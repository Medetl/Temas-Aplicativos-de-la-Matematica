```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t):
    return 10 - 10*np.exp(-5*t)
```

```python
x=np.linspace(0,10,50)
```

```python
with plt.xkcd():
    # Based on "Stove Ownership" from XKCD by Randall Munroe
    # https://xkcd.com/418/
    fig = plt.figure()
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines[['top', 'right']].set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim([0, 10])
    ax.set_ylim([0, 12])
    
    ax.annotate(
        'I(t) tiende a 10 Amperes',
        xy=(10, 10), arrowprops=dict(arrowstyle='->'), xytext=(5, 12))
    
    ax.plot(x,f(x), color='red')
    
    ax.set_xlabel('t')
    ax.set_ylabel('I(t)')

    #plt.savefig('grafica.png')
plt.show()
```
